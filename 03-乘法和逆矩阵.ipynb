### 1. 矩阵乘法

### (1). 形式化公式 

A, B, C为矩阵, 且: $A * B = C$ <br>
A行列数为: $(m*n)$, B行列数为: $(n*p)$ <br>
$A_{i,j}$为A矩阵第i行第j列的元素 <br>
$B_{i,j}$为B矩阵第i行第j列的元素 <br>
$C_{i,j}$为C矩阵第i行第j列的元素 <br>
则$C_{i,j}$计算公式为: <br>
$
\begin{equation*}
C_{i,j} = \sum_{k=1}^n A_{i,k} * B_{k,j}
\end{equation*}
$<br>
结果矩阵C的行列数为: ${m*p}$

由以上公式能进一步推导出: <br>
$
C_{i,j} = A.irow(i) * B.icol(j)
$

### (2). 行列规则

A为m行n列, B为n行p列 <br>
A的列数(n)要和B的行数相等 <br>
C为m行p列

$A_{m,n} * B_{n,p} = C_{m,p}$

### 2. 逆矩阵的求法

### (1) 原始方法

矩阵进行消元, 记录每次消元的矩阵: $E_n * ... * E_2 * E_1 * A $ <br>
$A^-1 = E_n * ... * E_2 * E_1 $

### (2) 高斯法

在A矩阵后并上一个I矩阵 $[A I]$ <br>
通过消元让原A部分位置变为I, 原I部分就记录了对应的消元过程, 即: $E_n * ... * E_2 * E_1$ <br>
操作过程见下等式: <br>
$ E * [A I] = [I X]$ <br>
$ E * A = I $ => $E = A^-1$ <br>
$E * I = X$ => $X = E$ <br>
=> $X = A^-1$

### 3. 矩阵乘法规则

### (1) 结合率

$(A * B) * C =  A * (B * C)$

### (2) 方阵的逆可交换

A为方阵, 且: $ A^-1 * A = I $ <br>
则: $ A * A^-1 = I $

(99). 未归类

$ A * A ^ -1 = I $ <br>
==> $ (A^-1)^T * A^T = I $
